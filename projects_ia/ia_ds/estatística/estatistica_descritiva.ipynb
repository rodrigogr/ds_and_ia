# Distribuição Normal (Gaussiana)
A distribuição normal, também conhecida como distribuição Gaussiana, é uma das distribuições estatísticas mais importantes e amplamente utilizadas na análise de dados. Ela é descrita por uma curva simétrica em forma de sino, que é definida por dois parâmetros:

<font color=green>Média (μ):</font> Representa o centro da distribuição, indicando o valor esperado ou o ponto mais provável dos dados.<br>
<font color=green>Desvio padrão (σ):</font> Mede a dispersão dos dados em relação à média. Valores > de 𝜎 resultam em uma curva mais achatada e espalhada, enquanto valores < geram uma curva mais estreita.

**Características Principais**
- É simétrica em torno da média (μ).
- A cauda da distribuição nunca toca o eixo-x, indicando que os valores extremos têm probabilidade muito baixa, mas nunca exatamente zero.
- A maior parte dos valores (cerca de 68%) está dentro de 1 desvio padrão da média (𝜇±𝜎), 95% dentro de 2 desvios padrões (𝜇±2𝜎), e 99,7% dentro de 3 desvios padrões (𝜇±3𝜎). Isso é conhecido como a regra empírica (Regra 68-95-99.7).

**Aplicações em análise de dados**
- Modelagem de fenômenos naturais: Muitos eventos no mundo real seguem aproximadamente uma distribuição normal (como alturas de pessoas, erros de medição e IQ).
- Inferência estatística: Vários testes estatísticos (como teste t) assumem que os dados seguem uma distribuição normal.
- Base para muitos testes estatísticos: Muitos testes estatísticos (teste t, ANOVA, etc.) assumem que os dados seguem uma distribuição normal.
- Facilidade de análise: A distribuição normal possui propriedades matemáticas bem definidas que facilitam a análise e inferência.

# Teorema do Limite Central (TLC)
É um dos conceitos mais fundamentais da estatística e tem implicações diretas para a análise de dados. Ele afirma que:

<font color= green>*Dado um conjunto de amostras aleatórias de uma população com qualquer tipo de distribuição (não necessariamente normal), a média das amostras tende a se aproximar de uma distribuição normal à medida que o tamanho da amostra aumenta.*</font>

Em outras palavras, independentemente da distribuição original dos dados, a distribuição das médias amostrais será aproximadamente normal para tamanhos de amostra suficientemente grandes.

**Implicações do TLC**
- Mesmo que a população original não seja normal, as médias amostrais serão distribuídas de forma aproximadamente normal se o tamanho da amostra for grande o suficiente (𝑛≥30 é uma regra prática comumente usada, embora possa variar dependendo da distribuição original).
- A média das médias amostrais será igual à média populacional (𝜇).
- O desvio padrão das médias amostrais é chamado de erro padrão , o que significa que a dispersão das médias diminui com o aumento do tamanho da amostra.

**Aplicações do TLC em análise de dados**
- Inferência estatística: O TLC permite que analistas apliquem testes baseados na distribuição normal, mesmo quando a distribuição populacional original não é normal.
- Construção de intervalos de confiança: Graças ao TLC, podemos usar a distribuição normal para fazer estimativas sobre parâmetros populacionais.
- Modelagem preditiva: Em situações de amostragem repetitiva, o TLC fundamenta muitas técnicas de modelagem.

# Resumo das relações entre distribuição Normal e TLC


- A distribuição normal é frequentemente usada como base teórica para modelagem de dados e testes estatísticos.
- O Teorema do Limite Central garante que, mesmo em situações onde os dados não sejam originalmente normais, as médias amostrais ainda podem ser tratadas como normais, permitindo a aplicação de ferramentas estatísticas baseadas nessa suposição.
- Ambos os conceitos são cruciais na análise de dados, pois permitem compreender padrões e realizar inferências confiáveis, mesmo em situações complexas.

# Medidas Posição, Dispersão e Forma
Medidas de posição, são importantes para resumir os dados que estamos lidando. As três principais medidas de posição são a <u>média, a mediana e a moda.</u> A média é a soma de todos os valores dividida pela quantidade de valores. A mediana é o valor do meio quando os dados estão ordenados e não é influenciada por valores extremos. A moda é o valor mais frequente em um conjunto de dados.

Essas medidas são ferramentas essenciais na estatística descritiva <font color=green>para resumir, descrever e interpretar conjuntos de dados.</font> Cada grupo oferece uma perspectiva diferente sobre o comportamento dos dados.
- Posição: Onde os dados estão localizados (ex.: média, mediana).
- Dispersão: Quão espalhados ou concentrados os dados estão (ex.: desvio padrão, amplitude).
- Forma: Como os dados estão distribuídos em termos de simetria e caudas (ex.: assimetria, curtose).

Essas três categorias são essenciais para descrever os dados: posição diz onde os valores estão, dispersão indica como estão espalhados, e forma revela detalhes sobre a distribuição. 

## Medias de Posição
<font color=green>Ponto central ou tendência central dos dados.</font>

As medidas de posição indicam onde os dados estão concentrados ou qual o valor típico de um conjunto de dados.
- Média: A soma de todos os valores dividida pelo número total de observações. Representa o "centro de gravidade" dos dados. Pode ser sensíveis a valores extremos (outliers).
- Mediana: O valor central quando os dados estão ordenados. É resistente a valores extremos (outliers).
- Moda: É o valor mais frequente em um conjunto de dados. Pode haver uma ou mais modas, ou o conjunto pode não ter uma moda. A moda é útil em dados categóricos ou quando se deseja identificar os valores mais frequentes.
- Percentis e Quartis: Indicadores da posição relativa de um valor em um conjunto de dados ordenados. Por exemplo, o 25º percentil (1º quartil) indica o ponto abaixo do qual estão 25% dos dados.

**Exemplo:**
Idade = {15, 18, 25, 25, 40, 55, 58, 60, 80}
- Média = 41,77
- Mediana = 40
- Moda = 25

## Medidas de Dispersão
<font color=green>Quão homogêneos ou heterogêneos são os dados</font>
As medidas de dispersão mostram o grau de variação ou dispersão dos dados em torno da média ou mediana.
- Amplitude (Range): Diferença entre o maior e o menor valor. Simples, mas sensível a outliers.
- Variância: É a média dos quadrados das diferenças entre cada valor e a média aritmética. Ela fornece uma ideia de quão distantes os valores estão da média, considerando o peso de cada diferença ao quadrado.
- Desvio Padrão: É a raiz quadrada da variância. Ele expressa a dispersão em termos da mesma unidade dos dados e é uma medida de dispersão mais comum.
- Coeficiente de Variação (CV): É o desvio padrão dividido pela média, expresso como porcentagem. Ele indica a variabilidade relativa dos dados em relação à média e é útil para comparar a dispersão entre conjuntos de dados diferentes.
- Amplitude Interquartil (IQR): Diferença entre o 3º quartil (Q3) e o 1º quartil (Q1). Mede a dispersão dos 50% centrais dos dados.

**Exemplo:**  
- Variância = 509,44
- Desvio Padrão = 22.58
- Coeficiente de Variação = 54,05% 
- Amplitude Interquartil (IQR)
  


## Cálculo Medidas De Dispersão
Calcularemos a variância em um conjunto de dados de altura e idade, assim nos ajudará a entender o quão dispersos estão os dados em relação à média. Em seguida, calcularemos o desvio padrão, que é a raiz quadrada da variância e nos dá uma medida mais intuitiva da dispersão dos dados. Por fim, calcularemos o coeficiente de variação, nos permite comparar a variabilidade de diferentes conjuntos de dados, dividindo o desvio padrão pela média e multiplicando por 100. No final concluíremos que o conjunto de dados de idade apresenta uma maior variabilidade em comparação ao conjunto de dados de altura.

```python
# Variância 
df_medidas['idade'].var()
```

```python
# Desvio Padrão
df_medidas['idade'].std()
```

```python
# Coeficiente de Variação
df_medidas['idade'].std() / df_medidas['idade'].mean() * 100
```

## Medidas de Forma
As medidas de forma avaliam a simetria e o achatamento da distribuição dos dados:

**Assimetria:** Indica o grau e a direção da distorção da distribuição em relação à média. Uma assimetria positiva significa que a cauda direita da distribuição é mais 
longa (os valores maiores estão mais espalhados), enquanto uma assimetria negativa significa que a cauda esquerda é mais longa. Ou seja:


- Se for 0, a distribuição é simétrica.
- Valores positivos indicam uma cauda mais longa à direita (assimetria positiva).
- Valores negativos indicam uma cauda mais longa à esquerda (assimetria negativa).
  
**Curtose:** Mede o pico ou a "pontuação" da distribuição, ou seja, o grau de achatamento ou "pontualidade" da distribuição. Uma curtose alta indica uma distribuição mais concentrada (pico mais agudo e caudas mais pesadas), enquanto uma curtose baixa indica uma distribuição mais achatada (pico menos agudo e caudas menos pesadas). Ou seja:


- Curtose > 0: Distribuição mais pontiaguda (leptocúrtica).<br>
- Curtose < 0: Distribuição mais achatada (platicúrtica).<br>
- Curtose = 0: Similar à normal (mesocúrtica).

Essas medidas ajudam a entender o formato e as características mais sutis da distribuição.

**Exemplo:** idade = {15, 18, 25, 25, 40, 55, 58, 60, 80}
- Assimetria = 0,3036 (Positiva = Moda < Mediana < Média)
- Curtose = -1,19 (Platicúrtica)

# Correlação
A correlação na estatística mede a relação entre duas variáveis, indicando se elas têm uma associação linear positiva (aumentam juntas), negativa (uma aumenta enquanto a outra diminui) ou nenhuma correlação. A importância para algoritmos de Machine Learning reside na capacidade de identificar padrões e relações entre variáveis. A correlação ajuda a selecionar características relevantes para os modelos, melhorando a precisão e interpretabilidade. Também permite ajustar modelos para prever com maior acurácia com base nas relações observadas nos dados.

Para medir a relação entre duas variáveis usaremos:
- **Coeficiente de Pearson:** que é usado para medir a correlação linear entre duas variáveis numéricas, variando de -1 a 1. 
- **Coeficiente de Spearman:** avalia uma relação monotônica, sendo útil quando os dados não têm uma relação linear clara ou quando as variáveis não são numericamente escalonáveis. 

**Exemplo:** 
Tempo de Serviço (meses) = {6, 10, 12, 18, 24, 30, 36, 50}<br>
Salário = {1800, 2000, 2400, 3000, 3600, 4300, 5100, 6000}
- Correlação de Pearson = 0,9944
- Correlação de Spearman = 1.00
- Coeficiente de Variação Tempo de Serviço = 64,03%
- Coeficiente de Variação Salário = 42,96%


## Cálculo Coeficiente Correlação
O método de correlação correlaciona todas as variáveis entre si. No entanto, podemos também correlacionar apenas duas variáveis específicas de forma explícita, escolhendo as variáveis desejadas. 

Usando o método de correlação de Pearson ou Spearman, podemos ver a correlação entre as variáveis selecionadas. Se não especificarmos o método, o Pandas assume o método de Pearson como padrão.

```python
# Correlação do Data Frame segundo Pearson
df_medidas.corr(method='pearson')
```
