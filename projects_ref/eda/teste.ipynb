```python
import pandas as pd

# Carregar os datasets
contracts_df = pd.read_csv('dados/churn_contracts.csv')
customers_df = pd.read_csv('dados/churn_customers.csv')
services_df = pd.read_csv('dados/churn_services.csv')

# Exibir as primeiras linhas de cada dataset para inspeção inicial
contracts_head = contracts_df.head(2)
customers_head = customers_df.head(2)
services_head = services_df.head(2)

```

```python
# Combinar os datasets usando 'customerID' como chave
merged_df = contracts_df.merge(customers_df, on='customerID').merge(services_df, on='customerID')

# Exibir as primeiras linhas do dataframe combinado
merged_head = merged_df.head()

# Verificar as informações do dataframe combinado
merged_info = merged_df.info()

# Calcular estatísticas descritivas
merged_describe = merged_df.describe(include='all')

# Verificar valores ausentes
missing_values = merged_df.isnull().sum()

merged_head, merged_info, merged_describe, missing_values
```
