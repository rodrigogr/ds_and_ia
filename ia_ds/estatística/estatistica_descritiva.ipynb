# Distribuio Normal (Gaussiana)
A distribuio normal, tambm conhecida como distribuio Gaussiana,  uma das distribuies estatsticas mais importantes e amplamente utilizadas na anlise de dados. Ela  descrita por uma curva simtrica em forma de sino, que  definida por dois parmetros:

<font color=green>Mdia ():</font> Representa o centro da distribuio, indicando o valor esperado ou o ponto mais provvel dos dados.<br>
<font color=green>Desvio padro ():</font> Mede a disperso dos dados em relao  mdia. Valores > de  resultam em uma curva mais achatada e espalhada, enquanto valores < geram uma curva mais estreita.

**Caractersticas Principais**
-  simtrica em torno da mdia ().
- A cauda da distribuio nunca toca o eixo-x, indicando que os valores extremos tm probabilidade muito baixa, mas nunca exatamente zero.
- A maior parte dos valores (cerca de 68%) est dentro de 1 desvio padro da mdia (), 95% dentro de 2 desvios padres (2), e 99,7% dentro de 3 desvios padres (3). Isso  conhecido como a regra emprica (Regra 68-95-99.7).

**Aplicaes em anlise de dados**
- Modelagem de fenmenos naturais: Muitos eventos no mundo real seguem aproximadamente uma distribuio normal (como alturas de pessoas, erros de medio e IQ).
- Inferncia estatstica: Vrios testes estatsticos (como teste t) assumem que os dados seguem uma distribuio normal.
- Base para muitos testes estatsticos: Muitos testes estatsticos (teste t, ANOVA, etc.) assumem que os dados seguem uma distribuio normal.
- Facilidade de anlise: A distribuio normal possui propriedades matemticas bem definidas que facilitam a anlise e inferncia.

# Teorema do Limite Central (TLC)
 um dos conceitos mais fundamentais da estatstica e tem implicaes diretas para a anlise de dados. Ele afirma que:

<font color= green>*Dado um conjunto de amostras aleatrias de uma populao com qualquer tipo de distribuio (no necessariamente normal), a mdia das amostras tende a se aproximar de uma distribuio normal  medida que o tamanho da amostra aumenta.*</font>

Em outras palavras, independentemente da distribuio original dos dados, a distribuio das mdias amostrais ser aproximadamente normal para tamanhos de amostra suficientemente grandes.

**Implicaes do TLC**
- Mesmo que a populao original no seja normal, as mdias amostrais sero distribudas de forma aproximadamente normal se o tamanho da amostra for grande o suficiente (30  uma regra prtica comumente usada, embora possa variar dependendo da distribuio original).
- A mdia das mdias amostrais ser igual  mdia populacional ().
- O desvio padro das mdias amostrais  chamado de erro padro , o que significa que a disperso das mdias diminui com o aumento do tamanho da amostra.

**Aplicaes do TLC em anlise de dados**
- Inferncia estatstica: O TLC permite que analistas apliquem testes baseados na distribuio normal, mesmo quando a distribuio populacional original no  normal.
- Construo de intervalos de confiana: Graas ao TLC, podemos usar a distribuio normal para fazer estimativas sobre parmetros populacionais.
- Modelagem preditiva: Em situaes de amostragem repetitiva, o TLC fundamenta muitas tcnicas de modelagem.

# Resumo das relaes entre distribuio Normal e TLC


- A distribuio normal  frequentemente usada como base terica para modelagem de dados e testes estatsticos.
- O Teorema do Limite Central garante que, mesmo em situaes onde os dados no sejam originalmente normais, as mdias amostrais ainda podem ser tratadas como normais, permitindo a aplicao de ferramentas estatsticas baseadas nessa suposio.
- Ambos os conceitos so cruciais na anlise de dados, pois permitem compreender padres e realizar inferncias confiveis, mesmo em situaes complexas.

# Medidas Posio, Disperso e Forma
Medidas de posio, so importantes para resumir os dados que estamos lidando. As trs principais medidas de posio so a <u>mdia, a mediana e a moda.</u> A mdia  a soma de todos os valores dividida pela quantidade de valores. A mediana  o valor do meio quando os dados esto ordenados e no  influenciada por valores extremos. A moda  o valor mais frequente em um conjunto de dados.

Essas medidas so ferramentas essenciais na estatstica descritiva <font color=green>para resumir, descrever e interpretar conjuntos de dados.</font> Cada grupo oferece uma perspectiva diferente sobre o comportamento dos dados.
- Posio: Onde os dados esto localizados (ex.: mdia, mediana).
- Disperso: Quo espalhados ou concentrados os dados esto (ex.: desvio padro, amplitude).
- Forma: Como os dados esto distribudos em termos de simetria e caudas (ex.: assimetria, curtose).

Essas trs categorias so essenciais para descrever os dados: posio diz onde os valores esto, disperso indica como esto espalhados, e forma revela detalhes sobre a distribuio. 

## Medias de Posio
<font color=green>Ponto central ou tendncia central dos dados.</font>

As medidas de posio indicam onde os dados esto concentrados ou qual o valor tpico de um conjunto de dados.
- Mdia: A soma de todos os valores dividida pelo nmero total de observaes. Representa o "centro de gravidade" dos dados. Pode ser sensveis a valores extremos (outliers).
- Mediana: O valor central quando os dados esto ordenados.  resistente a valores extremos (outliers).
- Moda:  o valor mais frequente em um conjunto de dados. Pode haver uma ou mais modas, ou o conjunto pode no ter uma moda. A moda  til em dados categricos ou quando se deseja identificar os valores mais frequentes.
- Percentis e Quartis: Indicadores da posio relativa de um valor em um conjunto de dados ordenados. Por exemplo, o 25 percentil (1 quartil) indica o ponto abaixo do qual esto 25% dos dados.

**Exemplo:**
Idade = {15, 18, 25, 25, 40, 55, 58, 60, 80}
- Mdia = 41,77
- Mediana = 40
- Moda = 25

## Medidas de Disperso
<font color=green>Quo homogneos ou heterogneos so os dados</font>
As medidas de disperso mostram o grau de variao ou disperso dos dados em torno da mdia ou mediana.
- Amplitude (Range): Diferena entre o maior e o menor valor. Simples, mas sensvel a outliers.
- Varincia:  a mdia dos quadrados das diferenas entre cada valor e a mdia aritmtica. Ela fornece uma ideia de quo distantes os valores esto da mdia, considerando o peso de cada diferena ao quadrado.
- Desvio Padro:  a raiz quadrada da varincia. Ele expressa a disperso em termos da mesma unidade dos dados e  uma medida de disperso mais comum.
- Coeficiente de Variao (CV):  o desvio padro dividido pela mdia, expresso como porcentagem. Ele indica a variabilidade relativa dos dados em relao  mdia e  til para comparar a disperso entre conjuntos de dados diferentes.
- Amplitude Interquartil (IQR): Diferena entre o 3 quartil (Q3) e o 1 quartil (Q1). Mede a disperso dos 50% centrais dos dados.

**Exemplo:**  
- Varincia = 509,44
- Desvio Padro = 22.58
- Coeficiente de Variao = 54,05% 
- Amplitude Interquartil (IQR)
  


## Clculo Medidas De Disperso

## Medidas de Forma
As medidas de forma avaliam a simetria e o achatamento da distribuio dos dados:

**Assimetria:** Indica o grau e a direo da distoro da distribuio em relao  mdia. Uma assimetria positiva significa que a cauda direita da distribuio  mais 
longa (os valores maiores esto mais espalhados), enquanto uma assimetria negativa significa que a cauda esquerda  mais longa. Ou seja:


- Se for 0, a distribuio  simtrica.
- Valores positivos indicam uma cauda mais longa  direita (assimetria positiva).
- Valores negativos indicam uma cauda mais longa  esquerda (assimetria negativa).
  
**Curtose:** Mede o pico ou a "pontuao" da distribuio, ou seja, o grau de achatamento ou "pontualidade" da distribuio. Uma curtose alta indica uma distribuio mais concentrada (pico mais agudo e caudas mais pesadas), enquanto uma curtose baixa indica uma distribuio mais achatada (pico menos agudo e caudas menos pesadas). Ou seja:


- Curtose > 0: Distribuio mais pontiaguda (leptocrtica).<br>
- Curtose < 0: Distribuio mais achatada (platicrtica).<br>
- Curtose = 0: Similar  normal (mesocrtica).

Essas medidas ajudam a entender o formato e as caractersticas mais sutis da distribuio.

**Exemplo:** idade = {15, 18, 25, 25, 40, 55, 58, 60, 80}
- Assimetria = 0,3036 (Positiva = Moda < Mediana < Mdia)
- Curtose = -1,19 (Platicrtica)

# Correlao
A correlao na estatstica mede a relao entre duas variveis, indicando se elas tm uma associao linear positiva (aumentam juntas), negativa (uma aumenta enquanto a outra diminui) ou nenhuma correlao. A importncia para algoritmos de Machine Learning reside na capacidade de identificar padres e relaes entre variveis. A correlao ajuda a selecionar caractersticas relevantes para os modelos, melhorando a preciso e interpretabilidade. Tambm permite ajustar modelos para prever com maior acurcia com base nas relaes observadas nos dados.

Para medir a relao entre duas variveis usaremos:
- **Coeficiente de Pearson:** que  usado para medir a correlao linear entre duas variveis numricas, variando de -1 a 1. 
- **Coeficiente de Spearman:** avalia uma relao monotnica, sendo til quando os dados no tm uma relao linear clara ou quando as variveis no so numericamente escalonveis. 

**Exemplo:** 
Tempo de Servio (meses) = {6, 10, 12, 18, 24, 30, 36, 50}<br>
Salrio = {1800, 2000, 2400, 3000, 3600, 4300, 5100, 6000}
- Correlao de Pearson = 0,9944
- Correlao de Spearman = 1.00
- Coeficiente de Variao Tempo de Servio = 64,03%
- Coeficiente de Variao Salrio = 42,96%
